```python
from magres.atoms import MagresAtoms
atoms = MagresAtoms.load_magres('8-HQ-ipc2-B_opt_magres.magres')
```

```python
for atom in atoms:
    print (atom, atom.ms.iso)
```

```python
for atom in atoms.species('B'):
    print (atom, "iso:",atom.ms.iso, 'aniso:',atom.ms.aniso)

```

```python
for atom in atoms.species('B'):
    print (atom, "sigma:\n",atom.ms.sigma)
    print()
```

```python
for atom in atoms.species('B'):
    print (atom, "Eigen Value:\n",atom.ms.evals,'\n' "iso: \n", atom.ms.iso)
    print("delta_ppm: ",(atom.ms.evals[2] - atom.ms.iso), '\n', "eta: ", (atom.ms.evals[0] -atom.ms.evals[1])/(atom.ms.evals[2] - atom.ms.iso))

 
```
