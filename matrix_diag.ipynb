```python
import numpy as np

shield_array = np.array([[-62.9830, 166.2263, 80.5663], #shielding tensor
                         [172.4778, 65.9098, -24.7315], 
                         [-8.2278, 25.6984, -51.8337]]) 

print(shield_array)
eigenvalues, eigenvectors = np.linalg.eig(shield_array) 
P = eigenvectors
P_inv = np.linalg.inv(eigenvectors)
D = np.diag(eigenvalues)
print("eigen vector: ", P)
print("inv of eigen vector: ", P_inv)
print("diagonal matrix of eigenvalues: ", D)
print(P@D@P_inv)
```

```python
import numpy as np
from scipy.linalg import eig, inv
M = np.matrix([[5, 2, 0], 
            [2, 5, 0], 
            [-3, 4, 6]]) 
eVals,eVecs = eig(M)    
print(M)
# print(eVecs)
D = np.zeros((3,3))
for i in range(0,len(eVals)):
    D[i,i] = eVals[i].real
Sinv = inv(eVecs)
print("eigen vector",eVecs)
print(D)
print(Sinv)
eVecs@D@(Sinv)
```
