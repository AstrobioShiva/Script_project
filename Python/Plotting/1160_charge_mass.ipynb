```python
from decimal import Decimal
def B(N, I,  a,  mu0):
    valueB = (mu0*N*I/a)*(4/5)**(1.5)
    return valueB
def cm(V, B,r):
    valuecm = (2*V)/(B*r)**2
    return valuecm
```

```python
e = 1.16*10e-19
m = 9.11*10e-31
N = 130
I = 1.18
a = 15e-2
m0 = 1.26e-6
V = 200
d = 10e-2
r = d/2
field = B(N,I,a,m0)
charge_mass = cm(V,field,r)
print('calculated charge to mass ratio',("{:e}".format(charge_mass)))
print('theoretical charge to mass', ("{:e}".format(e/m)))
```
