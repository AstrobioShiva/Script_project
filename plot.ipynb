```python
df = B_data_file()

import matplotlib.pyplot as plt
Fig, axes = plt.subplots(nrows=4, ncols=3,figsize=(16,12), sharex=True,constrained_layout=True)
title = ['Z rotation','Y rotation','X rotation']
for  i in range(3):
    axes[i,0].set_ylabel(r'$Frequency\ (Hz)$')
    axes[i+1, 0].set_ylabel(r'$Frequency\ (Hz)$')

    axes[0,i].set_title(title[i])
    axes[0,i].plot(df[i].angle,df[i].freq_1D, '-m', label = r'$|3/2>\leftrightarrow|1/2> transition$')
    axes[0,i].plot(df[i].angle,df[i].freq_2D, '-r',label = r'$|1/2>\leftrightarrow|-1/2> transition$')
    axes[0,i].plot(df[i].angle,df[i].freq_3D, '-b',label = r'$|-1/2>\leftrightarrow|-3/2> transition$')
    axes[0,i].legend(loc='best')
    axes[0,i].grid()

    axes[1,i].plot(df[i].angle,df[i].freq_sum, '-g', label = r'$3/2>\leftrightarrow|1/2> + |1/2>\leftrightarrow|-1/2>+ |-1/2>\leftrightarrow|-3/2>$')
    axes[1,i].legend(loc='best')
    axes[1,i].grid()

    axes[2,i].plot(df[i].angle,df[i].freq_diff, '-', label = r'$|3/2>\leftrightarrow|1/2> - |-1/2>\leftrightarrow|-3/2>$')
    axes[2,i].set_xlim(left=0)
    axes[2,i].legend(loc='best') 
    axes[2,i].grid()





    axes[3,i].plot(df[i].angle,df[i].acs, label = r'$H_{Q, ACS_D}$')
    axes[3,i].plot(df[i].angle,df[i].csa, '-', color = 'olive',label = r'$H_{Q, CSA}$')
    # axes[3,i].plot(df[i].angle,df[i].acs + df[i].csa, label = r'$H_{Q, ACS} +H_{Q, CSA} $')
    axes[3,i].plot(df[i].angle,df[i].acs_anti,  '-', color='darkorange', label = r'$H_{Q, ACS_L}$')
    axes[3,i].set_xlabel(r'$Angle\ (degrees)$')
    axes[3,i].legend(loc='best')
    axes[3,i].grid()
```
